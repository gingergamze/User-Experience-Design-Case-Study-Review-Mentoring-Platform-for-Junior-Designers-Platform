<li>THE PROJECT: Design a digital mentoring platform that matches junior and senior UX designer to discuss on a UX case study and collects UX tips from session notes</li>
    
<li>WEBSITES USED FOR SCRAPING: There were four suitable websites to harvest data about UX case studies. Casestudy.club, bootcamp.uxdesign.cc, uxdesign.cc and ux-station were the source of collection for the title and the synopsis of the UX case studies</li>

```python
import requests
from bs4 import BeautifulSoup #for webscraping
```

```python
abstractlist =[]
headers =[]
links= []
for i in range(1,25):  # this "for" makes the iteration on every page
    response=requests.get("https://www.casestudy.club/case-studies?292412a7_page={}".format(i)) #i becomes string with the format
    soup = BeautifulSoup( response.content,"html.parser")
    trial =soup.find_all("div",{"class":"cat-inner"})
    newheader = soup.find_all("h3",{"class":"cat-header"})
    
    headers += [header.text for header in newheader] # ever header in one page 
  
  
    
    refs = soup.find_all("a", {"class":"cat-link"})
    for link in refs:
        if link.text =="Direct":
            links.append(link["href"])
    
    
    for div in trial:
        result = ""
        newheader = div.find("h3",{"class":"cat-header"})
        result += newheader.text
        newlink="https://www.casestudy.club"+div.find("a")["href"]
        newsite=requests.get(newlink)
        newsoup = BeautifulSoup( newsite.content,"html.parser")
        newdiv = newsoup.find("div",{"class":"w-richtext"})
        if(newdiv.p == None):
            if not newdiv.text =="" :
                result += " " + newdiv.text
                
        else:
            if not newdiv.text =="" :
                result += " " + newdiv.p.text
        abstractlist.append(result)
              
```

```python
response=requests.get("https://uxdesign.cc/ux-case-studies/home")
soup2 = BeautifulSoup( response.content,"html.parser")
trial2 =soup2.find_all("div",{"class":"u-contentSansThin"})
newheader2 = soup2.find_all("h3")

uxcollective_headers=[]
uxcollective_synopsis=[]

for header in newheader2:
    uxcollective_headers.append(header.find("div").text)

for div in trial2:
    uxcollective_synopsis.append(div.text)
    
    
```

```python
with open("Medium.html", "r", encoding="utf-8") as infile:
    content=infile.read()
    
soup3 = BeautifulSoup(content,"html.parser")
trial3 =soup3.find_all("div",{"class":"js-trackPostPresentation"})


newheader3 = soup3.find_all("h3")

medium_headers=[]
medium_synopsis=[]

for header in newheader3:
    medium_headers.append(header.find("div").text)

for div in trial3:
    div2 = div.find("div", {"class": "u-marginTop7"})
    if not div2 == None:
        medium_synopsis.append(div2.text)
    else:
        medium_synopsis.append("")
   
```

```python
with open("bootcamp.html", "r", encoding="utf-8") as infile:
    content=infile.read()
    
soup4 = BeautifulSoup(content,"html.parser")
trial4 =soup4.find_all("div",{"class":"js-trackPostPresentation"})


newheader4 = soup4.find_all("h3")

bootcamp_headers=[]
bootcamp_synopsis=[]

for header in newheader4:
    bootcamp_headers.append(header.find("div").text)

    #appending the synopsis to the bootcamp_synopsis
for div in trial4:
    div2 = div.find("div", {"class": "u-marginTop7"})
    if not div2 == None:
        bootcamp_synopsis.append(div2.text)
    else:
        bootcamp_synopsis.append("")
```

```python
with open("uxdesign.html", "r", encoding="utf-8") as infile:
    content=infile.read()
    
    
soup4 = BeautifulSoup(content,"html.parser")
trial4 =soup4.find_all("div",{"class":"js-trackPostPresentation"})

newheader4 = soup4.find_all("h3")

uxdesign_headers=[]
uxdesign_synopsis=[]

for header in newheader4:
    uxdesign_headers.append(header.find("div").text)

for div in trial4:
    div2 = div.find("div", {"class": "u-marginTop7"})
    if not div2 == None:
        uxdesign_synopsis.append(div2.text)
    else:
        uxdesign_synopsis.append("")
```

```python
data=[]
index=0

for i in range(len(abstractlist)):
    data.append(abstractlist[i])
    
for i in range(len(bootcamp_headers)):
    data.append(bootcamp_headers[i]+" "+ bootcamp_synopsis[i])

for i in range(len(uxdesign_headers)):
    data.append(uxdesign_headers[i]+" "+ uxdesign_synopsis[i])

for i in range(len(medium_headers)):
    data.append(medium_headers[i]+" "+ medium_synopsis[i])


```

```python
response=requests.get("http://insightintechnology.com/top-100-ux-tips-for-designers/")
soup = BeautifulSoup( response.content,"html.parser")
tips=soup.find_all("li")
```

```python
tips[0].attrs["class"]
```

```python
notes=[]
for tip in tips:
    if not "class" in tip.attrs:
        notes.append(tip.text)
      
```
